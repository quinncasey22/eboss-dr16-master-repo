```python
def smooth_spec(lamb, fluxhat):

    #Smoothed spectra
    plt.style.use('grayscale')
    f, ax = plt.subplots(figsize=(12, 6)) 
    plt.xlabel('R-Frame Wavelength ($\AA$)')
    plt.ylabel('erg/s/$cm^2$/Ang')
    plt.title(plate)
    ax.step(lamb[np.where(snr > 1)], fluxhat[np.where(snr > 1)], c='k')
    
    return
```

```python
f1 = fits.open('/Users/quinncasey/Desktop/research/eboss/0.3-bin/0.3-spec-files-less/7906/spec-7906-57333-0843.fits')
f2 = fits.open('/Users/quinncasey/Desktop/research/eboss/0.3-bin/0.3-spec-files-less/4697/spec-4697-55660-0618.fits')
f3 = fits.open('/Users/quinncasey/Desktop/research/eboss/0.3-bin/0.3-spec-files-less/7378/spec-7378-56715-0693.fits')
f4 = fits.open('/Users/quinncasey/Desktop/research/eboss/0.3-bin/0.3-spec-files-less/8522/spec-8522-58193-0474.fits')
f5 = fits.open('/Users/quinncasey/Desktop/research/eboss/0.3-bin/0.3-spec-files-less/5166/spec-5166-56065-0372.fits')
```

```python
z = f1[2].data['z']
lamb = (10**f1[1].data['loglam']* u.AA)/(1+z)
flux = f1[1].data['flux']
err = np.sqrt(1/f1[1].data['ivar'])
plate = '7906-57333-0843'

snr = flux/err
flux = flux *10**-17* u.Unit('erg cm-2 s-1 AA-1')
fluxhat = scipy.signal.medfilt(flux)

smooth_spec(lamb, fluxhat)
```
